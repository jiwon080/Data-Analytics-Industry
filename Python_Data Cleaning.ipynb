```python
#subset the data
df1=df[['timestamp','company', 'title', 'location', 'yearsofexperience', 'yearsatcompany','totalyearlycompensation','basesalary','stockgrantvalue','gender','Education','Race']]
df1
```

```python
#remove all positions not related to DS
df1= df1[df1["title"].str.contains("Business Analyst|Data Scientist|Management Consultant|Software Engineer|Software Engineering Manager")]
df1
```

```python
missing = df1.isna().sum() 

print("The percentage of each variable's missing values are as follow:")
round(missing / len(df1) * 100, 2)
```

```python
#remove all locations outside of the united states
df2=df1[df1["location"].str.contains("AL|AK|AZ|AR|CA|CO|CT|DC|DE|FL|GA|HI|ID|IL|IN|IA|KS|KY|LA|ME|MD|MA|MI|MN|MS|MO|MT|NE|NV|NH|NJ|NM|NY|NC|ND|OH|OK|OR|PA|RI|SC|SD|TN|TX|UT|VT|VA|WA|WV|WI|WY")]
df2
```

```python
#drop remaining countries
new=df2[df2["location"].str.contains("Netherlands|Austria|France|Hong Kong|Spain|China|India|Australia|Poland|United Kingdom|Denmark|Nigeria|Colombia|Uruguay|Japan|Belarus|Italy|Latvia|Saudi Arabia|Switzerland|Brazil")==False]
new
```

```python
#remove "United States" from the end of locations
new['location'] = new['location'].map(lambda x: x.rstrip(', United States'))
```

```python
nan_value = float("NaN")
new.replace("", nan_value, inplace=True)
new.dropna(subset = ["timestamp","company","Education","Race","gender"], inplace=True)
```

```python
new['location'] = new['location'].replace('Kansas City, K','Kansas City, KS')
new['location'] = new['location'].replace('Olathe, K','Olathe, KS')
new['location'] = new['location'].replace('Wichita, K','Wichita, KS')
new['location'] = new['location'].replace('Overland Park, K','Overland Park, KS')
new['location'] = new['location'].replace('Oxford, M','Oxford, MS')
new['location'] = new['location'].replace('Mississippi State, M','Mississippi State, MS')
```

```python
# Take out commas
new.company = new.company.str.replace(',', '')
```

```python
# Remove hyphens
new.company = new.company.str.replace(' - ', ' ')
```

```python
# Remove white space
new.company = new.company.str.strip()
```
